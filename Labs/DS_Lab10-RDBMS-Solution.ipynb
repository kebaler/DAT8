```python
%matplotlib inline
import pandas as pd
import sqlite3
from pandas.io import sql
conn = sqlite3.connect('../data/enron.db')
import matplotlib.pyplot as plt
```

Lab - part 1
===================

Use matplotlib.pyplot.scatter (or plt.scatter for short) 
or ggplot : geom_point
to plot the length of the subjects against the recipient_count.

```python
subj_length = sql.read_sql("""SELECT LENGTH(subject) as subject_length,
                                        MAX(rno) as recipicient_count
                                        FROM MessageBase mb,
                                             RecipientBase rb
                                        WHERE mb.mid = rb.mid
                                        GROUP BY subject_length""", 
                                   conn, "subject_length")
subj_length.head()
```

```python
plt.scatter(subj_length.index, subj_length.values)
```

#### ...but something is lost when the data is queried in this manner:<br/><br/>    each subject length will only have the largest recipient count associated with it. <br/><br/> i.e. if there are two messages with the subject length of 10, one with a recipient count of 30 and the other with a recipient cound of 40; there will be only length-count pair in the data above <br/><br/> The following method will reflect all of the pairs:  

```python
msg_subj_length = sql.read_sql("""SELECT mid, LENGTH(subject) as subject_length
                                        FROM MessageBase mb""", 
                                   conn, "mid")
msg_subj_length.head()
```

```python
msg_r_count = sql.read_sql("""SELECT mid, MAX(rno) as recipicient_count
                                        FROM RecipientBase rb
                                        GROUP BY mid""", 
                                   conn, "mid")
msg_r_count.head()
```

```python
subj_len_r_count = msg_subj_length.join(msg_r_count)
subj_len_r_count.head()
```

```python
plt.scatter(subj_len_r_count['subject_length'],subj_len_r_count['recipicient_count'])
```

Lab - part 2
===================


Goal:
Plot the median subject length of the messages send grouping by week

```python
message_length = sql.read_sql("""SELECT mid, DATETIME(unix_time, 'unixepoch') AS time_sent, 
                                             LENGTH(subject) as subject_length, subject
                                        FROM MessageBase """, 
                                   conn, "mid")
message_length.head()
```

```python
group_by_week = message_length.time_sent.map(lambda dt: pd.Period(dt, freq='W'))
group_by_week.head(4)
```

```python
message_length.groupby(group_by_week).subject_length.median().plot()
```

```python
%%bash
# NOTE: use %%cmd in place of %%bash on Windows
# check that the database is located as expected
sqlite3 ../data/enron.db .tables
sqlite3 ../data/enron.db ".schema EmployeeBase"
sqlite3 ../data/enron.db ".schema RecipientBase"
sqlite3 ../data/enron.db ".schema MessageBase "
```

###Exploration: Compare length by seniority level by week

```python
message_length_level = sql.frame_query("""SELECT seniority, mid, DATETIME(unix_time, 'unixepoch') AS time_sent, 
                                             LENGTH(subject) as subject_length, subject
                                        FROM MessageBase, EmployeeBase 
                                        WHERE EmployeeBase.eid =  MessageBase.from_eid""", 
                                   conn, "mid")
message_length_level.tail()
```

```python
message_length_level[message_length_level["seniority"]=="Senior"].groupby(group_by_week).subject_length.median().plot(color='r')
message_length_level[message_length_level["seniority"]=="Junior"].groupby(group_by_week).subject_length.median().plot(color='b')
```

###Exploration: Compare mean recipient count by level by level sender over time

```python
message_level_level = sql.frame_query("""SELECT MessageBase.mid, eb_from.seniority seniority_from, eb_to.seniority seniority_to,  
                                                DATETIME(unix_time, 'unixepoch') AS time_sent, 
                                                COUNT(*) AS count
                                        FROM MessageBase, 
                                             EmployeeBase eb_from, 
                                             RecipientBase, 
                                             EmployeeBase eb_to
                                        WHERE MessageBase.from_eid = eb_from.eid
                                          AND MessageBase.mid = RecipientBase.mid
                                          AND RecipientBase.to_eid = eb_to.eid
                                          GROUP BY  MessageBase.mid, eb_from.seniority, eb_to.seniority,  DATETIME(unix_time, 'unixepoch')
                                        """, 
                                   conn, "mid")
message_level_level.head()
```

```python
message_level_level[message_level_level["seniority_from"]=="Senior"].groupby(group_by_week).mean().plot(color='r')
```

```python
message_level_level[message_level_level["seniority_from"]=="Junior"].groupby(group_by_week).mean().plot(color='r')
```

```python
message_level_level["from_to"]= \
        message_level_level.apply(lambda x: x['seniority_from'] + '-' + x['seniority_to'], axis=1)
    
message_level_level["week"] =  message_level_level.time_sent.map(lambda x: pd.Period(x, freq='W'))
#df['dt'] = pd.to_datetime(df.apply(lambda x: x['Date[G]'] + ' ' + x['Time[G]'], 1))
```

```python
message_level_level.head()
```

```python
message_level_level["week_end"]=message_level_level['week'].apply(lambda x: str(x).split('/')[0])
```

```python
mll_by_level_week=message_level_level.groupby(['from_to','week_end']).mean()
mll_by_level_week.head()
```
